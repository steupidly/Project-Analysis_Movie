```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from IPython.display import HTML
```

```python
df = pd.read_csv('movies_complete.csv')
```

```python
x =df.production_companies.str.split("|",expand=True).stack()
```

```python
x = x.to_frame().reset_index(level=1).drop(columns='level_1')
```

```python
x.value_counts()
```

```python
y= x.merge(df[['title', 'revenue_musd','release_date']],how='left',left_index=True,right_index=True)
```

```python
y =y.rename(columns={0:'production_com'})
```

```python
yx= y.groupby("production_com").agg({'title':'count','revenue_musd':'sum'}).nlargest(10,'title')
```

```python
yx
```

```python
# most active
yx.sort_values(by='title',ascending=False)
```

```python
#most succesfull
yx.sort_values(by='revenue_musd',ascending=False)
```

```python
gen = df.genres.str.split("|",expand = True).stack().reset_index(level=1, drop=True).to_frame()
```

```python
gen.columns= ['genre']
```

```python
gen = gen.merge(df[['title','revenue_musd','release_date']],how='left', left_index=True,right_index=True)
```

```python
date = df.release_date.between('1980-01-01','1989-12-31')
```

```python
gen
```

```python
gen_date = gen.loc[date,['genre','title','revenue_musd']]
```

```python
gen_date = gen_date.groupby('genre').agg({'title':'count','revenue_musd':'sum'})
```

```python
#most populer/active
gen_date.sort_values(by='title' ,ascending=False)
```

```python
# most succesfull
gen_date.sort_values(by='revenue_musd' ,ascending=False)
```

```python
date_90 = df.release_date.between('1990-01-01','1999-12-31')
```

```python
gen_date90 = gen.loc[date_90,['genre','title','revenue_musd']]
```

```python
gen_date90=gen_date90.groupby('genre').agg({'title':'count','revenue_musd':'sum'})
```

```python
gen_date90.sort_values(by='title',ascending=False)
```

```python
gen_date90.sort_values(by='revenue_musd',ascending=False)
```
